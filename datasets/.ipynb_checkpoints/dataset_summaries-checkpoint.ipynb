```python
import pandas as pd
import numpy as np
import glob
from itertools import groupby
from tqdm import tqdm 

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
filepaths = [x for x in glob.glob("./processed/**", recursive=True) if x.endswith(".csv") and "_all" not in x]
```

```python
def summarize_dataset(filepath):

    dataset_dict = {}
    df = pd.read_csv(filepath)
    filepath_split = filepath.split("/")

    dataset_dict['filepath'] = filepath
    dataset_dict['dataset'] = filepath_split[2]
    dataset_dict['sub_dataset'] = "/".join(filepath.split("/")[3:-1])
    dataset_dict['ts_name'] = filepath_split[-1][:-4]

    try:
        try:
            _ = int(df['timestamp'].values[0])
            dataset_dict['has_timestamp'] = False
        except:
            dataset_dict['has_timestamp'] = True

        if dataset_dict['has_timestamp']:
            df['ts'] = pd.to_datetime(df['timestamp'])
            dataset_dict['ts_range'] = df['ts'].values[-1] - df['ts'].values[0]
            dataset_dict['mean_ts_diff'] = df['ts'].diff().mean()
            dataset_dict['std_ts_diff'] = df['ts'].diff().std()


        consolidated_anomalies = [(k,len(list(v))) for k,v in groupby(df['is_anomaly']) if k == 1]
        dataset_dict['num_point_anomalies'] = len([x for x in consolidated_anomalies if x[1] == 1])
        dataset_dict['num_context_anomalies'] = len([x for x in consolidated_anomalies if x[1] != 1])
        dataset_dict['total_anomalies'] = len(consolidated_anomalies)
        dataset_dict['context_anomaly_lengths'] = [x[1] for x in consolidated_anomalies if x[1] != 1]
        dataset_dict['mean_context_anomaly_length'] = np.mean(dataset_dict['context_anomaly_lengths'])
        dataset_dict['total_context_anomaly_length'] = np.sum(dataset_dict['context_anomaly_lengths'])
        dataset_dict['total_anomaly_length'] = sum([x[1] for x in consolidated_anomalies])
        dataset_dict['dataset_length'] = len(df)
        
        #Out of dist 
        std_value = df['value'].std()
        mean_value = df['value'].mean()
        df['std_dist_from_mean'] = np.abs((df['value'] - df['value'].mean())/df['value'].std())
        
        dataset_dict['4std_OOD'] = np.sum(df['std_dist_from_mean'] >= 4)
        dataset_dict['3std_OOD'] = np.sum(df['std_dist_from_mean'] >= 3)

        dataset_dict['4std_OOD_anomalies'] = np.sum(df.query("is_anomaly == 1")['std_dist_from_mean'] >= 4)
        dataset_dict['3std_OOD_anomalies'] = np.sum(df.query("is_anomaly == 1")['std_dist_from_mean'] >= 3)
        
        dataset_dict['avg_std_dist_anomaly'] = df.query("is_anomaly == 1")['std_dist_from_mean'].mean()
        dataset_dict['avg_std_dist_notanomaly'] = df.query("is_anomaly != 1")['std_dist_from_mean'].mean()
        
        dataset_dict['med_std_dist_anomaly'] = df.query("is_anomaly == 1")['std_dist_from_mean'].median()
        dataset_dict['med_std_dist_notanomaly'] = df.query("is_anomaly != 1")['std_dist_from_mean'].median()
    
    except:
        print(filepath)
        
        
    return dataset_dict
```

```python
dataset_dicts = []
for filepath in tqdm(filepaths): 
    dataset_dicts.append(summarize_dataset(filepath))
```

```python
df_datasets = pd.DataFrame(dataset_dicts)
```

```python
from fbprophet import Prophet
```

```python
df = pd.read_csv("./processed/NASA/D-14_nasa.csv")
```

```python
m = Prophet()
df['ds'] = pd.to_datetime(df['timestamp'])
df['y'] = df['value']
```

```python
m.fit(df)
```

```python
plt.plot(df.ds, df.value)
plt.plot(df.ds, df.is_anomaly)
```

```python
plt.plot(df.index, df.value)
plt.plot(df.index, df.is_anomaly)
```

```python
plt.plot(df.index, df.value)
plt.plot(df.index, df.is_anomaly-3)
```

```python
df_predict = m.predict(df)
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df.value)
plt.plot(df_predict.index, df_predict.yhat)

plt.fill_between(df_predict.index, df_predict['yhat_lower'], df_predict['yhat_upper'], alpha=0.5)
plt.plot(df.index, df.is_anomaly/10, c="red")
```
